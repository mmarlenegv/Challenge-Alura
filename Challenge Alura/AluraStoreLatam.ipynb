### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
##Ingreso total por cada tienda
Ingreso_total_tienda1 = sum(tienda['Precio'])
Ingreso_total_tienda2 = sum(tienda2['Precio'])
Ingreso_total_tienda3 = sum(tienda3['Precio'])
Ingreso_total_tienda4 = sum(tienda4['Precio'])
print(f'Ingreso total de la tienda 1 es : {Ingreso_total_tienda1}')
print(f'Ingreso total de la tienda 2 es : {Ingreso_total_tienda2}')
print(f'Ingreso total de la tienda 3 es : {Ingreso_total_tienda3}')
print(f'Ingreso total de la tienda 4 es : {Ingreso_total_tienda4}')
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print(f'La cantidad de producto por categoría de la tienda 1: \n{ventas_por_categoria_tienda1}')
```

```python
print(f'La cantidad de producto por categoría de la tienda 2: \n{ventas_por_categoria_tienda2}')
```

```python
print(f'La cantidad de producto por categoría de la tienda 3: \n{ventas_por_categoria_tienda3}')
```

```python
print(f'La cantidad de producto por categoría de la tienda 4: \n{ventas_por_categoria_tienda4}')
```

```python
##Ventas agrupadas por Categoría
venta1 = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio',ascending=False)
venta2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio',ascending=False)
venta3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio',ascending=False)
venta4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio',ascending=False)
```

```python
print(f'Las ventas agrupadas por categoría de la tienda 1: \n{venta1}')
```

```python
print(f'Las ventas agrupadas por categoría de la tienda 2: \n{venta2}')
```

```python
print(f'Las ventas agrupadas por categoría de la tienda 3: \n{venta3}')
```

```python
print(f'Las ventas agrupadas por categoría de la tienda 4: \n{venta4}')
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda1 = tienda['Calificación'].mean().round(2)
promedio_calificacion_tienda2 = tienda2['Calificación'].mean().round(2)
promedio_calificacion_tienda3 = tienda3['Calificación'].mean().round(2)
promedio_calificacion_tienda4 = tienda4['Calificación'].mean().round(2)
```

```python
print(f' El promedio de calificación de la tienda 1 es: {promedio_calificacion_tienda1}')
print(f' El promedio de calificación de la tienda 2 es: {promedio_calificacion_tienda2}')
print(f' El promedio de calificación de la tienda 3 es: {promedio_calificacion_tienda3}')
print(f' El promedio de calificación de la tienda 4 es: {promedio_calificacion_tienda4}')
```
